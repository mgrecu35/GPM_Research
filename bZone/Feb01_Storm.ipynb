```python
import netCDF4 as nc
import numpy as np
import matplotlib.pyplot as plt
import matplotlib as mpl
import cartopy.crs as ccrs

fname='2A.GPM.DPR.V9-20211125.20200201-S122944-E140217.033678.V07A.HDF5'
import os
os.system('pwd')

def read_plot(fname,n1,n2):
    fhDPR=nc.Dataset(fname)
    lonD=fhDPR['FS/Longitude'][:]
    latD=fhDPR['FS/Latitude'][:]
    zMeasured=fhDPR['FS/PRE/zFactorMeasured'][:]
    bzd=fhDPR['FS/VER/binZeroDeg'][:]
    pType=fhDPR['FS/CSF/typePrecip'][:]
    heightZero=fhDPR['FS/VER/heightZeroDeg'][:]
    bcf=fhDPR['FS/PRE/binClutterFreeBottom'][:]
    plt.pcolormesh(lonD[n1:n2,24],range(176),zMeasured[n1:n2,24,:,0].T,vmin=0,vmax=50,cmap='jet')
    plt.ylim(175,100)
    return lonD[n1:n2,:],latD[n1:n2,:],zMeasured[n1:n2,:,:,0],pType[n1:n2,:],heightZero[n1:n2,:],bzd[n1:n2,:],bcf[n1:n2,:]
```

```python
fname='2A.GPM.DPR.V9-20211125.20200201-S140218-E153450.033679.V07A.HDF5'
n1,n2=2850,3000
lon,lat,zM,pType,hZero,bzd,bcf=read_plot(fname,n1,n2)
```
